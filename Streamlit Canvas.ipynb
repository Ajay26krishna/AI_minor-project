```python
%%writefile app.py
from tensorflow import keras
import numpy as np
import streamlit as st
import cv2
model = keras.models.load_model('MNIST_model.hdf5')
st.title('Digit Recognizer')
from streamlit_drawable_canvas import st_canvas

# Create a canvas component
st.write('Original Image')
canvas_result = st_canvas(stroke_width=10,stroke_color='#ff0000',background_color="#000000",
    height=150,width=150,drawing_mode='freedraw')

if canvas_result.image_data is not None:
    img = cv2.resize(canvas_result.image_data.astype('uint8'),(28,28))
    rescaled = cv2.resize(img,(150,150), interpolation = cv2.INTER_NEAREST)
    st.write('Rescaled Image')
    st.image(rescaled)

if st.button('PREDICT'):
    test_x = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    value = model.predict(test_x.reshape(1,28,28))
    st.write(f'Predicted Number: {np.argmax(value,axis=1)}')
    st.bar_chart(value[0])
```
